```python
def high_low(x):
    a=[]
    for i in x:
        if i<=x.quantile(0.25):
            a.append('Very Low')
        elif (i>=x.quantile(0.25)) and (i<x.quantile(0.5)):
            a.append('Low')
        elif (i>=x.quantile(0.5)) and (i<x.quantile(0.75)):
            a.append('High')
        else:
            a.append('Very High')
    return a
            
```

```python
homicide['Homicide_Level'] = high_low(homicide['Homicide_Rate'])
```

```python
crude_suicide_rate_total = crude_suicide_rate[crude_suicide_rate['Sex'] == ' Both sexes']
crude_suicide_rate_total.reset_index(drop = True, inplace = True)
crude_suicide_rate_male = crude_suicide_rate[crude_suicide_rate['Sex'] == ' Male']
crude_suicide_rate_male.reset_index(drop = True, inplace = True)
crude_suicide_rate_female = crude_suicide_rate[crude_suicide_rate['Sex'] == ' Female']
crude_suicide_rate_female.reset_index(drop = True, inplace = True)
crude_suicide_rate_total.head()
```

```python
crude_suicide_rate_female.head()
```

```python
crude_suicide_rate_female.columns = ['Country'] + ['crude_suicide_'+ i.strip() +'_female' for i in crude_suicide_rate_female.columns[1:]]
crude_suicide_rate_male.columns = ['Country'] + ['crude_suicide_'+ i.strip() +'_male' for i in crude_suicide_rate_male.columns[1:]]
crude_suicide_rate_total.columns = ['Country'] + ['crude_suicide_'+ i.strip() for i in crude_suicide_rate_total.columns[1:]]
```

```python
mental_health_policy.columns = ['Country', 'Stand_alone_law_for_mental_health', 'Stand_alone_policy_for_mental_health' ]
```

```python
def clean(x):
    a=[]
    for i in x:
        i = i.title()
        i = i.strip()
        a.append(i)
    return a
        
```

```python
mental_health_policy['Stand_alone_law_for_mental_health'] = clean(mental_health_policy['Stand_alone_law_for_mental_health'])
mental_health_policy['Stand_alone_policy_for_mental_health'] = clean(mental_health_policy['Stand_alone_policy_for_mental_health'])
```

```python
dataframes = [homicide, age_standardize_suicide_rate, crime, crude_suicide, facilities, happiness, 
              human_resource, mental_health_policy]
dataframes_1 = [age_standardize_suicide_rate, crude_suicide, happiness]
```

```python
for i in dataframes:
    i['Country'] = [j.lower().strip() for j in i['Country']]
```

```python
from functools import reduce
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Country'],
                                            how='inner'), dataframes_1)
```

```python
for i in [homicide, crime, facilities, human_resource, mental_health_policy]:
    df_merged = df_merged.merge(i, how = 'left', on = 'Country')
```
